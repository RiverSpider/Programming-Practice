```python
#1
import pandas as pd

data = {"name": ["Strom, Mrs. Wilhelm (Elna Matilda Persson)", "Navratil, Mr. Michel (Louis M Hoffman)", "Minahan, Miss. Daisy E"],
        "age": [29, 36.5, 33],
        "sex": ["female", "male", "female"]}

my_df = pd.DataFrame(data)

my_df
```

```python
#2
titanic_df = pd.read_csv('../content/sample_data/titanic_csv.csv')
titanic_df
```

```python
titanic_df = titanic_df.rename (
    columns = {
        'PassengerId': 'passengerid', 'Survived': 'survived', 'Pclass': 'pclass', 'Name': 'name', 'Sex': 'sex',
       'Age': 'age', 'SibSp': 'sibsp', 'Parch': 'parch', 'Ticket': 'ticket',
       'Fare': 'fare', 'Cabin': 'cabin', 'Embarked': 'embarked'
    }
)
```

```python
#3
gist_df = pd.read_csv('https://gist.githubusercontent.com/zaryanezrya/8b4ef51c707cb16d5e88a44dc00a1bb2/raw/41230f49c6268e072dbf102672f670be256922ab/gistfile1.txt')
gist_df
```

```python
#4
combined_df = pd.concat([titanic_df, gist_df])

combined_df = combined_df.drop_duplicates()
combined_df
```

```python
#5
combined_df = combined_df.set_index('passengerid')

combined_df = combined_df.sort_index()
combined_df
```

```python
#6
combined_df.info()

combined_df.describe()
```

```python
#7
combined_df.iloc[0], combined_df.iloc[2] = combined_df.iloc[2].copy(), combined_df.iloc[0].copy()
```

```python
#8
combined_df['sex'] = combined_df['sex'].map({'female': 'f', 'male': 'm'})
combined_df
```

```python
#9
grouped = combined_df.groupby('ticket').count()

result = grouped[grouped['name'] >= 6]['name']
print(result)
```

```python
#10
merged_df = combined_df.merge(my_df, on='name', how='inner')

common_cabins = merged_df['cabin'].unique()

people_in_common_cabins = combined_df[combined_df['cabin'].isin(common_cabins)]
people_in_common_cabins[['name', 'cabin']]
```

```python
#11
combined_df['birthyear'] = 1912  - combined_df['age']
combined_df
```

```python
#12
combined_df['сompanion'] = combined_df.groupby('cabin')['cabin'].transform('count') - 1

combined_df['сompanion'].fillna(0, inplace=True)
combined_df
```

```python
#13
indices = combined_df[combined_df.index == 1].index

if len(indices) > 1:
    combined_df.iloc[0], combined_df.iloc[indices[0]] = combined_df.iloc[indices[0]].copy(), combined_df.iloc[0].copy()
```

```python
#14
combined_df.to_csv('../content/sample_data/output.csv')
```

```python
#15
combined_df['fare'] = pd.to_numeric(combined_df['fare'], errors='coerce')

combined_df = combined_df.dropna(subset=['fare'])

top_10_passengers = combined_df.sort_values(by='fare', ascending=False).head(10)
top_10_passengers[['name', 'fare']]
```

```python
#16
survival_count = combined_df.groupby(['sex', 'survived']).size().unstack()

survival_count.rename(columns={0: 'Not Survived', 1: 'Survived'}, inplace=True)

survival_count
```

```python
#17
survival_count_by_class = combined_df.groupby(['pclass', 'survived']).size().unstack()

# Переименовываем столбцы для удобства
survival_count_by_class.rename(columns={0: 'Not Survived', 1: 'Survived'}, inplace=True)
survival_count_by_class
```
