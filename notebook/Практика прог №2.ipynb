```python
import numpy as np
import sympy as sp
from scipy.misc import derivative
from scipy.integrate import quad
from scipy.optimize import minimize_scalar

#y = sin(ln(x)), x0 = 4, a = 3, b = 8
```

```python
#1
def f(x):
    return np.sin(np.log(x))

x0 = 4
a = 3
b = 8

# Первая производная
dy_dx1 = derivative(f, x0, dx=1e-6)
print(f"Первая производная в точке x0={x0} равна {dy_dx1:.5f}")

# Вторая производная
dy_dx2 = derivative(f, x0, dx=1e-6, n=2)
print(f"Вторая производная в точке x0={x0} равна {dy_dx2:.5f}")
```

```python
#2
x = sp.symbols('x')
y = sp.sin(sp.log(x))

dydx_1 = sp.diff(y, x, 1)
print("Первая производная:", dydx_1)
dydx_2 = sp.diff(y, x, 2)
print("Вторая производная:", dydx_2)
```

```python
#3
integral, error = quad(f, a, b)

print(f"Определенный интеграл: {integral:.5f}")
print(f"Погрешность: {error}")
```

```python
#4
inte = sp.integrate(y, x)
print("Неопределенный интеграл:", inte)
```

```python
#5
from scipy.optimize import minimize

x1, x2 = sp.symbols('x1 x2')
f = (x1-3)**2 + x2
grad_f = [sp.diff(f, x) for x in [x1, x2]]

#ограничения
cons = [{'type': 'ineq', 'fun':
         lambda x: -2*x[0] + 3*x[1] - 4},
        {'type': 'ineq', 'fun':
         lambda x: x[0]},
        {'type': 'ineq', 'fun':
         lambda x: x[1]}]
x0 = [0,0]

#мин функция
res = minimize(lambda x: f.subs([(x1, x[0]), (x2, x[1])]), x0, method='SLSQP', constraints=cons)
print(res)
```
