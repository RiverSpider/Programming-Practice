```python
import numpy as np

def printmat(mat):
  for row in mat:
    for el in row:
      print(el, end=' ')
    print()

def printvec(vec):
  for element in np.nditer(vec):
    print(element, end=' ')
```

```python

#1
my_array = np.arange(10, 70, 2)
print('my_array: ')
printvec(my_array)
```

```python
#2
A = np.reshape(my_array, (6,5))
A = np.transpose(A)
print('A: ')
printmat(A)
```

```python
#3
A = A * 2.5
A[0, :] -= 3
print('A: ')
printmat(A)
```

```python
#4
B = np.random.randint(0, 10, (6, 3))
print('B: ')
printmat(B)
```

```python
#5
a = np.sum(A, axis=1)
print('a: ')
printvec(a)
b = np.sum(B, axis=0)
print('b: ')
printvec(b)
```

```python
#6
C = np.matmul(A,B)
print('C: ')
printmat(C)
```

```python
#7
A = np.delete(A, 2, axis = 1)
print('A: ')
printmat(A)
columns = np.random.randint(10,20, size=(B.shape[0], 3))
B = np.hstack((B, columns))
print('B: ')
printmat(B)
```

```python
#8
detA = np.linalg.det(A)
detB = np.linalg.det(B)
print(detA)
print(detB)
if(detA != 0):
  revA = np.linalg.inv(A)
  print('revA: ')
  printmat(revA)
if(detB != 0):
  revB = np.linalg.inv(B)
  print('revB: ')
  printmat(revB)
```

```python
#9
A = np.linalg.matrix_power(A, 6)
print('A: ')
printmat(A)
B = np.linalg.matrix_power(B, 14)
print('B: ')
printmat(B)
```

```python
#10
#Ex = g, тк a,b заняты
E = np.array([[4, -3.4, 1, 1.8],
             [-2, 8, 0 ,-9],
             [2, -2.5, -7, 4],
             [1, -1, 5, 1]])
g = np.array([6, -4.8, 10, 3])
x = np.linalg.solve(E, g)
print('x: ')
printvec(x)
```
