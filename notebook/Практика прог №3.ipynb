```python
import numpy as np

def printmat(mat):
  for row in mat:
    for el in row:
      print(round(el, 1), end=' ')
    print()
```

```python
#1
matrix = np.array([[4, -3.4, 1, 1.8],
                   [-2, 8, 0 ,-9],
                   [2, -2.5, -7, 4],
                   [1, -1, 5, 1]])
```

```python
#2
import scipy.linalg as la

p, l, u = la.lu(matrix)
print("LUP разложение матрицы:\n")
print("P: ")
printmat(p)
print("\nL: ")
printmat(l)
print("\nU: ")
printmat(u)
```

```python
#3
det = la.det(l) * la.det(u)

print(f"Определитель матрицы: {det:.3f} ")
inverse_p = la.inv(p)
det_p = la.det(inverse_p)

print(f"Определитель обратной матрицы P: {det_p:.3f}")
```

```python
#4
uniform_sample = np.random.randint(-10, 11, size=100)
normal_sample = np.random.normal(-10, 11, size=100)
```

```python
#5
from scipy import stats

uniform_mean = np.mean(uniform_sample)
normal_mean = np.mean(normal_sample)

uniform_mode = stats.mode(uniform_sample)
normal_mode = stats.mode(normal_sample)

uniform_median = np.median(uniform_sample)
normal_median = np.median(normal_sample)

uniform_min = np.min(uniform_sample)
normal_min = np.min(normal_sample)

uniform_max = np.max(uniform_sample)
normal_max = np.max(normal_sample)

uniform_std = np.std(uniform_sample)
normal_std = np.std(normal_sample)

print("Характеристики выборки с равномерным распределением:")
print(f"Среднее: {uniform_mean:.3f}")
print(f"Мода: {uniform_mode[0]:.3f}")
print(f"Медиана: {uniform_median:.3f}")
print(f"Минимум: {uniform_min:.3f}")
print(f"Максимум: {uniform_max:.3f}")
print(f"Стандартное отклонение: {uniform_std:.3f}\n")

print("Характеристики выборки с нормальным распределением:")
print(f"Среднее: {normal_mean:.3f}")
print(f"Мода: {normal_mode[0]:.3f}")
print(f"Медиана: {normal_median:.3f}")
print(f"Минимум: {normal_min:.3f}")
print(f"Максимум: {normal_max:.3f}")
```

```python
#6
from scipy.stats import chi2_contingency

num_bins = 10 #количество интервалов
observed_uniform_ids, _ = np.histogram(uniform_sample, bins=np.arange(num_bins + 1)) # количетсво наблюдений в каждом интервале
observed_normal_ids, _ = np.histogram(normal_sample, bins=np.arange(num_bins + 1))

table = np.vstack([observed_uniform_ids, observed_normal_ids])

chi2_statistic, p_value, _, _ = chi2_contingency(table)
print(f"Значение статистики критерия хи-квадрат: {chi2_statistic:.3f}")
print(f"p-value для нулевой гипотезы: \"Распределение выборки не равномерное\":  {p_value:.3f}")
```
