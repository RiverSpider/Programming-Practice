```python
import numpy as np
```

```python
#1
a = 3
b = 8
num_points = 10

x = np.linspace(a, b, num_points)

y = np.sin(np.log(x))

for i in range(num_points):
    print(f"x = {x[i]:.3f}, y = {y[i]:.3f}")
```

```python
#2
import matplotlib.pyplot as plt

plt.plot(x, y)

plt.title("График функции y = sin(ln(x))")
plt.xlabel("x")
plt.ylabel("y")

#Установка размеров шрифта для названия и подписей
plt.title("График функции y = sin(ln(x)),", fontsize=12)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=10)

#Добавление легенды
plt.legend(["y = sin(ln(x)),"])

plt.show()
```

```python
#3
plt.scatter(x, y, marker='o', color=(1, 0.5, 0.2))

plt.title("Точечный график")

plt.xlabel("x")
plt.ylabel("y")

#Добавление сетки
plt.grid(color='gray', linestyle='--', linewidth=0.5)

plt.show()
```

```python
#4
uniform_sample = np.random.randint(-10, 11, size=100)
normal_sample = np.random.normal(-10, 11, size=100)

#Построение гистограммы для выборки из равномерного распределения
plt.hist(uniform_sample, bins=20, color='blue', alpha=0.7)
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Гистограмма выборки из равномерного распределения')

#Отображение гистограммы для выборки из нормального распределения
plt.figure()
plt.hist(normal_sample, bins=30, color='green', alpha=0.7)
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Гистограмма выборки из нормального распределения')

plt.show()
```

```python
#5
sample = np.random.randint(low=0, high=5, size=50)

# Подсчет вхождений каждого числа в выборке
counts = np.bincount(sample)

# Данные для диаграммы
labels = ['1', '2', '3', '4']
colors = ['blue', 'green', 'orange', 'red']

# Построение круговой диаграммы
plt.figure(figsize=(6, 6))
plt.pie(counts[:-1], labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Круговая диаграмма")
plt.show()

# Построение столбчатой диаграммы
plt.figure(figsize=(6, 4))
plt.bar(labels, counts[:-1], color=colors)
plt.xlabel("Числа")
plt.ylabel("Количество")
plt.title("Столбчатая диаграмма")
plt.show()
```

```python
#6
from mpl_toolkits.mplot3d import Axes3D

def f(x, y):
    return np.sin(np.log(x))

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Построение трехмерного графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='cool')

plt.show()
```

```python
#7
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

fig.suptitle('Набор Графиков', fontsize=16)

axs[0, 0].plot(x, y)
axs[0, 0].set_title('Линейный график')
axs[0, 0].set_facecolor('lightgray')

axs[0, 1].scatter(x, y, marker='o', color=(1, 0.5, 0.2))
axs[0, 1].set_title('Точечный график')
axs[0, 1].set_facecolor('lightblue')

axs[1, 0].pie(counts[:-1], labels=labels, colors=colors, autopct='%1.1f%%')
axs[1, 0].set_title('Круговая диаграмма')
axs[1, 0].set_facecolor('lightgreen')

axs[1, 1] = fig.add_subplot(224, projection='3d')
axs[1, 1].plot_surface(X, Y, Z, cmap='cool')
axs[1, 1].set_title('3Д график')
axs[1, 1].set_facecolor('lightyellow')

plt.show()
```
