```python
#5
import matplotlib.pyplot as plt
import seaborn as sns

# a. Гистограмма распределения числового признака released_year
plt.figure(figsize=(10, 6))
sns.histplot(df['released_year'], bins=20, kde=True)
plt.title('Гистограмма распределения числового признака released_year')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.savefig('../content/histogram.png')
plt.show()
```

```python
# b. Диаграмма "ящик с усами" для числового признака in_spotify_playlists
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['in_spotify_playlists'])
plt.title('Диаграмма "ящик с усами" для числового признака in_spotify_playlists')
plt.xlabel('Значение')
plt.savefig('boxplot.png')
plt.show()
```

```python
# c. Круговая диаграмма для номинативного признака released_year
plt.figure(figsize=(8, 8))
df['released_year'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Круговая диаграмма для номинативного признака released_year')
plt.ylabel('')
plt.savefig('pie_chart.png')
plt.show()
```

```python
# d. Тепловая карта с корреляцией между признаками
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Тепловая карта с корреляцией')
plt.savefig('heatmap.png')
plt.show()
```

```python
# e. Диаграмма countplot с группировкой по двум номинативным признакам key mode
plt.figure(figsize=(12, 8))
sns.countplot(x='key', hue='mode', data=df)
plt.title('Диаграмма countplot с группировкой по двум номинативным признакам key mode')
plt.savefig('countplot.png')
plt.show()
```

```python
#6
if df.isnull().any().any():
    for column in df.columns:
        if df[column].dtype == 'int64':
            # Заполните пропуски медианой для целых чисел
            df[column].fillna(df[column].median(), inplace=True)
        elif df[column].dtype == 'float64':
            # Заполните пропуски средним значением для действительных чисел
            df[column].fillna(df[column].mean(), inplace=True)
        else:
            # Заполните пропуски модой для остальных типов данных
            df[column].fillna(df[column].mode().iloc[0], inplace=True)
df.isna().sum()
```

```python
#7
from scipy import stats
sample_size = 200
sample = df['released_year'].sample(n=sample_size, random_state=42)

statistic, p_value = stats.normaltest(sample)

if p_value < 0.05:
    print("Выборка не является нормально распределенной.")
else:
    print("Выборка может быть нормально распределенной.")
```
